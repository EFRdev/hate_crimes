## Cleaning the columns: 
- NOMCOMAR in the .json file "Comarcas"
- NOMMUNI in the .json file "Municipis"
- NOMPROV in the .json file "Provincies"

```python
comarcas_gdf = gpd.read_file("data/divisions-administratives-v2r1-20250101/divisions-administratives-v2r1-comarques-5000-20250101.json")
municipis_gdf = gpd.read_file("/Users/enrique/code/EFRdev/00 - Post-Bootcamp/Hate_Crimes/data/divisions-administratives-v2r1-20250101/divisions-administratives-v2r1-municipis-5000-20250101.json")
provincies_gdf = gpd.read_file("/Users/enrique/code/EFRdev/00 - Post-Bootcamp/Hate_Crimes/data/divisions-administratives-v2r1-20250101/divisions-administratives-v2r1-provincies-5000-20250101.json")
```

```python

def cleaning_text(text):

    text = unicodedata.normalize('NFKD', text).encode('ASCII', 'ignore').decode('utf-8')
    text = text.lower()
    text = text.replace("municipi de ", "")
    text = text.replace("resta municipis abp ", "")
    text = text.replace("abp ", "")
    text = text.replace("l'", "")
    text = text.replace("-", " ")
    text = text.replace("'", "")
    text = text.replace("pla de estany", "pla de lestany")
    text = text.strip()

    return text
```

```python
comarcas_gdf['NOMCOMAR_clean'] = comarcas_gdf['NOMCOMAR'].apply(cleaning_text)
comarcas_gdf
```

```python
municipis_gdf['NOMMUNI_clean'] = municipis_gdf['NOMMUNI'].apply(cleaning_text)
municipis_gdf.head()
```

```python
provincies_gdf['NOMPROV_clean'] = provincies_gdf['NOMPROV'].apply(cleaning_text)
provincies_gdf.head()
```
